```python
import requests
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
pages = []
base_url = 'https://www.house.kg/kupit-kvartiru'

for page_num in range(1, 6):
    url = f'{base_url}?page={page_num}' if page_num > 1 else base_url
    page_content = requests.get(url).content
    pages.append(page_content)
```

```python
soups = []
for page in pages:
    soup = BeautifulSoup(page, 'lxml')
    soups.append(soup)
```

### All apart

```python
apartments = soup.find_all('div',class_ = 'listing')# + soup2.find_all('div',class_ = 'listing') + soup3.find_all('div',class_ = 'listing') + soup4.find_all('div',class_ = 'listing')+ soup5.find_all('div',class_ = 'listing')+soup6.find_all('div',class_ = 'listing')+soup7.find_all('div',class_ = 'listing')+soup8.find_all('div',class_ = 'listing')+soup9.find_all('div',class_ = 'listing')+soup10.find_all('div',class_ = 'listing')
```

```python
apartments
```

```python
room = []
area = []
floor = []
address = []
price_in_usd = []
price_in_som = []

for apartment in apartments:
    main_info = apartment.find('p', class_='title').text.strip().split(',')
    
    room_1 = main_info[0].split('-')
    room.append(int(room_1[0]))

    area_1 = main_info[1]
    area.append(int(area_1[1]))

    floor_1 = main_info[2].split(' ')
    if floor_1[1] == 'цоколь':  
        floor.append(0) 
    else:
        floor.append(int(floor_1[1]))

    address.append(apartment.find('div', class_='address').text.strip().split(','))

    price_usd = apartment.find('div', class_='price').text.split(' ')
    price_in_usd.append(int(price_usd[1] + price_usd[2]))

    price_som = apartment.find('div', class_='price-addition').text.strip().split(' ')
    price_in_som.append(int(price_som[0] + price_som[1] + price_som[2]))
```

```python
df = pd.DataFrame(
        {
            'room':room,
            'area':area,
            'floor':floor,
            'address':address,
            'usd_price':price_in_usd,
            'som_price':price_in_som
        }
    )
```
